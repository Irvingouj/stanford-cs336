## What is a language model?

A language model assigns probabilities to sequences of tokens:

$$P(w_1, w_2, \dots, w_n) = \prod_{i=1}^{n} P(w_i \mid w_1, \dots, w_{i-1})$$

## Tokenization

Before we can model language, we need to break text into tokens.
Common approaches:
- **BPE** (Byte-Pair Encoding) — used by GPT-2/3/4
- **WordPiece** — used by BERT
- **SentencePiece** — used by LLaMA, T5
- **tiktoken** — OpenAI's fast BPE tokenizer

```python
import tiktoken

# Let's explore OpenAI's tokenizer
enc = tiktoken.get_encoding("cl100k_base")  # GPT-4 tokenizer

text = "Stanford CS336: Language Modeling from Scratch"
tokens = enc.encode(text)
print(f"Text: {text}")
print(f"Tokens: {tokens}")
print(f"Decoded: {[enc.decode_single_token_bytes(t) for t in tokens]}")
```

```python
# Vocabulary size
print(f"Vocabulary size: {enc.n_vocab}")
```

Implement the BPE algorithm:
1. Start with characters as tokens
2. Count adjacent token pairs
3. Merge the most frequent pair
4. Repeat until desired vocabulary size

```python
# TODO: implement BPE tokenizer here
from collections import Counter

def get_stats(ids):
    """Count adjacent pairs in a list of token ids."""
    counts = Counter()
    for pair in zip(ids, ids[1:]):
        counts[pair] += 1
    return counts

def merge(ids, pair, new_id):
    """Replace all occurrences of pair with new_id."""
    new_ids = []
    i = 0
    while i < len(ids):
        if i < len(ids) - 1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
            new_ids.append(new_id)
            i += 2
        else:
            new_ids.append(ids[i])
            i += 1
    return new_ids

# Test on a small corpus
corpus = "aaabdaaabac"
print(f"Corpus: {corpus}")
print(f"Characters: {sorted(set(corpus))}")
```
